```python
import tenserflow as tf
from tenserflow import keras
from keras import layers

train_ds=keras.utils.image_dataset_from_directory("dataset/train",image_size=(64,64),batch_size=32,label_model="categorical")

train_per=keras.Sequential([layers.Rescaling(1/255,),layers.RandomFlip("horizontal_and_vertical"),
layers.RandomRotation(0.2),
layers.RandomZoom(0.2),])

test_ds=keras.utils.image_dataset_from_directory("dataset/test",image_size=(64,64),batch_size=32,label_model="categorical"))

test_per=keras.Sequential([layers.Rescaling(1/255,)])
```

```python
AUTOTUNE=tf.data.AUTOTUNE
train_ds=train_ds.map(lambda x,y:train_per(x,training=True),y),num_parallel_calls=AUTOTUNE.prefetch(buffer_size=AUTOTUNE)
teset_ds=train_ds.map(lambda x,y:train_per(x,training=False),y),num_parallel_calls=AUTOTUNE.prefetch(buffer_size=AUTOTUNE)
```

```python
cnn=tf.keras.Sequatial()
cnn.add(keras.layers.Conv2D(filter_size=32,kernal_size=3,activation="relu",input_size=([64,64,3])))
cnn.add(keras.layers.MaxPool2D(pool_size=2,strids=2))
                            
```
